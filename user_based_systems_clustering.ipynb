```python
import numpy as np
import math
import pandas as pd
from sklearn.cluster import KMeans, SpectralClustering
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import accuracy_score, silhouette_score
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
'''
Reading in the data and merging the datasets
'''
```

```python
meta = pd.read_json('meta_Video_Games.json.gz',compression = 'gzip', lines = True,encoding = 'utf-8')
reviews = pd.read_json('Video_Games_5.json.gz', nrows = 10000, compression = 'gzip', lines = True, encoding = 'utf-8')
reviews.head(10)
```

```python
meta.head()
```

```python
reviews.head(30)
```

```python
review_df = reviews.pivot_table(index = 'reviewerID', columns = 'asin', values = 'overall').fillna(0)
review_df.head
```

```python
#tsne of the data set
tsne = TSNE(learning_rate = 'auto', init = 'random', perplexity = 5, random_state = 42)
xtsne = tsne.fit_transform(review_df)
plt.figure(figsize=(10,8))
plt.scatter(xtsne[:,0], xtsne[:,1], s=8, alpha=0.7)
plt.title('User Based Scatter')
plt.show()
```

```python
'''
SVD/COSINE_SIMILARITY IMPLEMENTATION
'''
```

```python
U, S, Vh = np.linalg.svd(review_df, full_matrices = False)
sim = []
for i in range(1, Vh.shape[1]):
    j = cosine_similarity(Vh[:,0].reshape(-1,1), Vh[:,i].reshape(-1,1))
    sim.append(np.linalg.norm(j))
plt.plot(range(1,Vh.shape[1]), sim)
plt.title('Testing Different SVD Matrices for Cosine Similarity')
plt.xlabel('Vh Arrays')
plt.ylabel('Similarity')
plt.show()
```

```python
'''
It seems like doing cosine similarity for user-based does not work because there are a greater number of users in comparison to products
'''
```

```python
'''
NEAREST NEIGHBORHOOD IMPLEMENTATION
'''
```

```python
distances = []
for i in range(1,10):
    nn = NearestNeighbors(n_neighbors = i, algorithm='auto')
    nn.fit(review_df)
    dist, ind = nn.kneighbors(review_df)
    distances.append(dist.mean())
plt.plot(range(1,10), distances)
plt.title('Finding N for Nearest Neighborhood')
plt.xlabel('Number of Neighbors')
plt.ylabel('Distance')
plt.show()
```

```python
#choosing 3 neighbors
test_points = np.random.choice(review_df.index, size=10, replace=False)
nn = NearestNeighbors(n_neighbors = 3, algorithm='auto')
nn.fit(review_df)
for points in test_points:
    neighborhood = nn.kneighbors(np.array(review_df.loc[points]).reshape(1,-1), n_neighbors = 3, return_distance = False)
    neighbor_reviews = pd.DataFrame(columns = review_df.columns)
    for neighbor in neighborhood[0]:
        neighbor_reviews = neighbor_reviews.append(review_df.iloc[neighbor])
    mean = neighbor_reviews.mean()
    i = 0
    mean = mean.sort_values(ascending=False)
    #print(mean.index[0])
    #recommendations = []
    print('recommendations for ', reviews.loc[reviews['reviewerID'] == points, 'reviewerName'].unique(), ':')
    while i < 5 and mean[i] != 0:
        if review_df.loc[points][mean.index[i]] == 0:
            print(meta.loc[meta['asin'] == mean.index[i], 'title'].unique())
        i+=1
```

```python
'''
SPECTRAL CLUSTERING IMPLEMENTATION
'''
```

```python
review_array = review_df.to_numpy()
print(review_array[0])
silhouette = []
for i in range(1,10):
    spectral = SpectralClustering(n_clusters = i, assign_labels = 'discretize', random_state=42)
    labels = spectral.fit_predict(review_array)
    silhouette.append(silhouette_score(review_array, labels))
    print(i)
plt.plot(range(1,10), silhouette)
plt.show()
```

```python
'''
KMEANS IMPLEMENTATION
'''
```

```python
inert = []
for i in range(1,30):
    kmean = KMeans(n_clusters = i, random_state = 42)
    kmean.fit(review_df)
    inert.append(kmean.inertia_)
    
plt.plot(range(1,30), inert)
plt.title('KMeans Cluster Analysis')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
test_points = np.random.choice(review_df.index, size=10, replace=False)
kmean = KMeans(n_clusters = 25, random_state = 42)
review_df['clusters'] = kmean.fit_predict(review_df)
```

```python
test_points = np.random.choice(review_df.index, size=10, replace=False)
for points in test_points:
    cluster = review_df.loc[points, 'clusters']
    print(cluster)
    group = review_df.loc[review_df['clusters'] == cluster]
    group = group.drop(['clusters'], axis=1)
    mean = group.mean()
    mean = mean.sort_values(ascending=False)
    i=0
    print('recommendations for ', reviews.loc[reviews['reviewerID'] == points, 'reviewerName'].unique(), ':')
    while i < 5 and mean[i] != 0:
        if review_df.loc[points][mean.index[i]] == 0:
            print(meta.loc[meta['asin'] == mean.index[i], 'title'].unique())
        i+=1    
```

```python
'''
DBSCAN
'''
```

```python
silhouette = []
review_array = review_df.values
for e in range(2,28):
    dbs = DBSCAN(eps = e, min_samples=1)
    labels = dbs.fit_predict(review_array)
    silhouette.append(silhouette_score(review_array, labels))
    print(e)
    
plt.plot(range(2,28), silhouette)
plt.title('DBSCAN Analysis')
plt.xlabel('eps')
plt.ylabel('Inertia')
plt.show()
```

```python
#dbscan of 4 works the best because it goes negative before going back up suggesting overfitting
dbs = DBSCAN(eps = 4, min_samples = 1)
review_df['scan'] = dbs.fit_predict(review_df)
```
